```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x, y, learning_rate, iterations):
    """
        Perform gradient descent for multivariate linear regression.
    Args:
        X: Feature matrix of shape (m, n) where m is the number of samples, and n is the number of features.
        y: Target values of shape (m, 1).
        learning_rate: Step size for gradient descent.
        iterations: Number of iterations to run gradient descent.
    Returns:
        weights: Coefficients for the features (including intercept).
        cost_history: List of cost values for each iteration.
    """
    
    m, n = x.shape
    x = np.hstack((np.ones((m, 1)), x))  # Add bias term (column of ones)
    weights = np.zeros((n + 1, 1))  # Initialize weights (including intercept)
    cost_history = []  
    
    for _ in range(iterations):
        # Calculate predictions
        y_pred = np.dot(x, weights)
        # Calculate error
        error = y_pred - y
        # Compute gradient
        gradient = (1 / m) * np.dot(x.T, error)
        # Update weights
        weights -= learning_rate * gradient
        # Calculate and store cost
        cost = (1 / (2 * m)) * np.sum(error ** 2)
        cost_history.append(cost)

    return weights, cost_history

```

```python
x = np.array([[2104, 3],
              [1600, 3],
              [2400, 3],
              [1416, 5],
              [3000, 4]])
# Target: House prices (in $1000s)
y = np.array([[400],
              [330],
              [369],
              [532],
              [940]])

```

```python
learning_rate = 0.0000001  # Small learning rate for large feature values
iterations = 1000
weights, cost_history = gradient_descent(x, y, learning_rate, iterations)
```

```python
print("Final Weights (including intercept):")
print(weights)
```

```python
plt.plot(range(len(cost_history)), cost_history, color='blue')
plt.title("Cost Reduction Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```
